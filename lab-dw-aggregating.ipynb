```python
import pandas as pd

url="https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
marketing_customer_analysis=pd.read_csv(url)
marketing_customer_analysis
marketing_customer_analysis.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\lab-dw-data-aggregation-and-filtering\marketing_customer_analysis.csv")
```

```python
#cleaning data


marketing_customer_analysis.columns=marketing_customer_analysis.columns.str.lower()
marketing_customer_analysis.columns=marketing_customer_analysis.columns.str.replace(" ", "_").str.lower()

marketing_customer_analysis
marketing_customer_analysis.to_csv(r"D:\04_IronHack_DA\IRONHACK_LABS_PROJECTS\lab-dw-data-aggregation-and-filtering\marketing_customer_analysis.csv")

marketing_customer_analysis.head(5)
#marketing_customer_analysis.columns

marketing_customer_analysis.columns
```

```python
#data frame with total claim amount < 1000


df_total_claim_amount = marketing_customer_analysis[marketing_customer_analysis["total_claim_amount"]<1000]
df_total_claim_amount.head()
```

```python
# data frame with total claim amount < 1000 and response = Yes

df_total_claim_amount_grouped=df_total_claim_amount.groupby("response")
df_total_claim_amount_grouped_yes = df_total_claim_amount_grouped.get_group("Yes")
df_total_claim_amount_grouped_yes
```

```python
# average monthly_premium and/or customer lifetime value by policy_type and gender


marketing_customer_analysis_response=marketing_customer_analysis.groupby("response")
marketing_customer_analysis_response_yes=marketing_customer_analysis_response.get_group("Yes")
marketing_customer_analysis_response_yes

display(marketing_customer_analysis_response_yes.groupby(["gender", "policy_type"])["customer_lifetime_value"].mean())
display(marketing_customer_analysis_response_yes.groupby(["gender", "policy_type"])["monthly_premium_auto"].mean())
```

```python
# compare these insights to total_claim_amount patterns, and discuss which segments appear most profitable or low-risk for the company.

display(marketing_customer_analysis_response_yes.groupby(["gender", "policy_type"])["total_claim_amount"].mean())
```

```python
#  Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

df_customers=marketing_customer_analysis[["customer","state","policy_type"]]

df_customers
```

```python
state_customers = df_customers.groupby("state")["customer"].nunique()
state_customers_over_500 = state_customers[state_customers > 500]
display(state_customers)
display(state_customers_over_500)

# All the sates have unique customers above 500
```

```python
# Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

df_max_min = marketing_customer_analysis.groupby("gender")
print(round(df_max_min.get_group("M")["customer_lifetime_value"].mean()),2)
print(round(df_max_min.get_group("F")["customer_lifetime_value"].mean(),2))

# Females on average have more lifetime value than males.
```
